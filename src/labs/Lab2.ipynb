```python
## Q4

# I wouldn't expect the quantile transformation to improve the decision tree model. 
# A quantile transformation doesn't depend on the scale or difference in values 
# but rather the ordering. With decision trees, because the split occurs if a value 
# crosses a certain threshold, the ordering is already preserved
```

```python
## Q5
from sklearn.preprocessing import QuantileTransformer
from sklearn.neighbors import KNeighborsClassifier

ic_homes_q5 = pd.read_csv("https://remiller1450.github.io/data/IowaCityHomeSales.csv")

ic_homes_q5['over_assessed'] = np.where(ic_homes_q5['assessed'] > ic_homes_q5['sale.amount'], "yes", "no")
ic_q5_train, ic_q5_test = train_test_split(ic_homes_q5, test_size=0.2, random_state=7)

ic_q5_train_y = ic_q5_train['over_assessed']
ic_q5_train_X = ic_q5_train.select_dtypes("number").drop(['sale.amount', 'assessed'], axis=1)
ic_q5_test_y = ic_q5_test['over_assessed']
ic_q5_test_X = ic_q5_test.select_dtypes("number").drop(['sale.amount', 'assessed'], axis=1)

knn_model = KNeighborsClassifier(n_neighbors=10, weights='distance', p=2)
qt = QuantileTransformer()
X_train_transformed = qt.fit_transform(ic_q5_train_X)
X_test_transformed = qt.transform(ic_q5_test_X)
knn_model.fit(X_train_transformed, ic_q5_train_y)

accuracy = knn_model.score(X_test_transformed, ic_q5_test_y)
print(f"Test set accuracy: {accuracy}")
```
